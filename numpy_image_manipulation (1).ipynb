```python
print(type(np_image), '\t', np_image.shape)
```

```python
np_image[:3, :3, 0]                      # RED values for top-left 3x3 pixels
```

```python
plt.imshow(np_image)
```

## invert colors

```python
np_image_inverted = np_image[:, :, ::-1] 
```

```python
np_image_inverted.shape
```

```python
np_image_inverted[:3, :3, 0]
```

```python
np_image_inverted[0, 0, :]
```

```python
plt.imshow(np_image_inverted)
```

## horizontal flip

```python
np_image_flipped = np_image[:, ::-1] 
```

```python
np_image_flipped.shape
```

```python
np_image_flipped[:3, :3, 0]
```

```python
np_image_flipped[0, 0, :]
```

## greyscale

```python
np_image_greyscale = np.mean(np_image, axis=-1)
```

```python
np_image_greyscale.shape
```

```python
np_image_greyscale[:3, :3]
```

```python
plt.imshow(np_image_greyscale, cmap='Greys')
```

## b&w

```python
np_image_blackandwhite = np.copy(np_image_greyscale)
np_image_blackandwhite[np_image_blackandwhite < 128] = 0
np_image_blackandwhite[np_image_blackandwhite >= 128] = 255
```

```python
np_image_blackandwhite.shape
```

```python
np_image_blackandwhite[:3, :3]
```

```python
plt.imshow(np_image_blackandwhite, cmap='Greys')
```
